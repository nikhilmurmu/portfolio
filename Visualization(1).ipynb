```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/user/HS400/kb01548/Downloads/archive(4) (1)/hotel_bookings.csv'
hotel_bookings = pd.read_csv(file_path)

# Display initial info
print("Initial DataFrame Info:")
print(hotel_bookings.info())

# Remove rows with null values
hotel_bookings_cleaned = hotel_bookings.dropna()
print(f"\nRows after removing null values: {len(hotel_bookings_cleaned)}")

# Remove duplicate rows
hotel_bookings_cleaned = hotel_bookings_cleaned.drop_duplicates()
print(f"Rows after removing duplicate rows: {len(hotel_bookings_cleaned)}")

# Display basic statistics of the cleaned data
print("\nBasic Statistics of Cleaned Data:")
print(hotel_bookings_cleaned.describe())

# Set Seaborn theme for visualizations
sns.set_theme(style="whitegrid")

# Visualization 1: Cancellations by Hotel Type
plt.figure(figsize=(8, 6))
sns.countplot(data=hotel_bookings_cleaned, x='hotel', hue='is_canceled')
plt.title('Cancellations by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.legend(title='Is Canceled', labels=['No', 'Yes'])
plt.show()

# Visualization 2: Average lead time by cancellation status
plt.figure(figsize=(8, 6))
sns.barplot(data=hotel_bookings_cleaned, x='is_canceled', y='lead_time', ci=None, palette="muted")
plt.title('Average Lead Time by Cancellation Status')
plt.xlabel('Is Canceled (0=No, 1=Yes)')
plt.ylabel('Average Lead Time')
plt.show()

# Visualization 3: Relationship between special requests and cancellations
plt.figure(figsize=(8, 6))
sns.boxplot(data=hotel_bookings_cleaned, x='is_canceled', y='total_of_special_requests', palette="Set2")
plt.title('Special Requests vs Cancellations')
plt.xlabel('Is Canceled (0=No, 1=Yes)')
plt.ylabel('Total Special Requests')
plt.show()

# Visualization 4: Cancellation rate by customer type
plt.figure(figsize=(8, 6))
customer_cancellation = hotel_bookings_cleaned.groupby('customer_type')['is_canceled'].mean().reset_index()
sns.barplot(data=customer_cancellation, x='customer_type', y='is_canceled', palette="pastel")
plt.title('Cancellation Rate by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Cancellation Rate')
plt.show()

# Visualization 5: Distribution of ADR (Average Daily Rate) by hotel type
plt.figure(figsize=(8, 6))
sns.boxplot(data=hotel_bookings_cleaned, x='hotel', y='adr', palette="coolwarm")
plt.title('Distribution of Average Daily Rate by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.show()
```

```python
# Visualization for Distribution Channel
plt.figure(figsize=(10, 6))
sns.countplot(data=hotel_bookings_cleaned, x='distribution_channel', hue='is_canceled')
plt.title('Cancellations by Distribution Channel')
plt.xlabel('Distribution Channel')
plt.ylabel('Number of Bookings')
plt.legend(title='Is Canceled', labels=['No', 'Yes'])
plt.xticks(rotation=45)
plt.show()

# Visualization for Market Segment
plt.figure(figsize=(10, 6))
sns.countplot(data=hotel_bookings_cleaned, x='market_segment', hue='is_canceled')
plt.title('Cancellations by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Number of Bookings')
plt.legend(title='Is Canceled', labels=['No', 'Yes'])
plt.xticks(rotation=45)
plt.show()

# Visualization for Required Car Parking Spaces
plt.figure(figsize=(8, 6))
sns.barplot(data=hotel_bookings_cleaned, x='required_car_parking_spaces', y='is_canceled', ci=None, palette="viridis")
plt.title('Cancellation Rate by Required Car Parking Spaces')
plt.xlabel('Required Car Parking Spaces')
plt.ylabel('Cancellation Rate')
plt.show()

# Visualization for Agent
top_agents = hotel_bookings_cleaned['agent'].value_counts().head(10).index
filtered_data = hotel_bookings_cleaned[hotel_bookings_cleaned['agent'].isin(top_agents)]
plt.figure(figsize=(10, 6))
sns.countplot(data=filtered_data, x='agent', hue='is_canceled')
plt.title('Cancellations by Top 10 Agents')
plt.xlabel('Agent ID')
plt.ylabel('Number of Bookings')
plt.legend(title='Is Canceled', labels=['No', 'Yes'])
plt.show()

# Visualization for Customer Type
plt.figure(figsize=(8, 6))
sns.countplot(data=hotel_bookings_cleaned, x='customer_type', hue='is_canceled', palette="cool")
plt.title('Cancellations by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Number of Bookings')
plt.legend(title='Is Canceled', labels=['No', 'Yes'])
plt.show()
```

```python
# Distribution of Days in Waiting List without KDE
plt.figure(figsize=(10, 6))
sns.histplot(data=hotel_bookings_cleaned, x='days_in_waiting_list', bins=30, hue='is_canceled', palette="husl", multiple="stack")
plt.title('Distribution of Days in Waiting List')
plt.xlabel('Days in Waiting List')
plt.ylabel('Count')
plt.legend(title='Is Canceled', labels=['No', 'Yes'])
plt.show()


```
